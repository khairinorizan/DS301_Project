```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import os
import pandas as pd

#data_path = os.path.join("/content/sample_data", "emails.csv")
df = pd.read_csv("emails.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

newData = df.drop(['Email No.', 'Prediction'], axis = 1)

x_train, x_test, y_train, y_test = train_test_split(newData, df['Prediction'], test_size=0.2, random_state=23)

logisticRegr = LogisticRegression(max_iter=300, solver='sag')
logisticRegr.fit(x_train, y_train)

print("Accuracy:",metrics.accuracy_score(y_test, logisticRegr.predict(x_test)))
print("%0.3f" % logisticRegr.score(x_train, y_train))
```
